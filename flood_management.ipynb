# For the better analysis, we are processing our entire data in range(0, 1)

```python
from sklearn import preprocessing
minmax = preprocessing.MinMaxScaler(feature_range=(0,1))
minmax.fit(x).transform(x)
```

# KNN Algorithm Starts here 


<!-- Converting our Entire data in two parts, i.e., Training and Testing Data in the ratio 4:1 -->

```python
from sklearn import model_selection,neighbors
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state = 0)
```

# Creating Knn object

```python
clf = neighbors.KNeighborsClassifier()
knn_clf = clf.fit(x_train,y_train)
```

# Predicting the Flood and storing the result in y_predict

```python
y_predict = knn_clf.predict(x_test)
print('Predicted chances of flood: ')
print(y_predict)
```

# Printing the Accurcy

```python
print(knn_clf.score(x_test,y_test))
```

# Actual Floods

```python
print("Actual values of floods:")
print(y_test)
```

# Cross Checking the Output

```python
from sklearn.model_selection import cross_val_score
knn_accuracy = cross_val_score(knn_clf,x_test,y_test,cv=3,scoring='accuracy',n_jobs=-1)
```

# Printing Accuracy

```python
accuracy_in_prediction = knn_accuracy.mean()
print(round(accuracy_in_prediction *100,2), "%")
```
